```python
import pandas as pd

recipes = pd.read_csv("budgetbytes_recipes.csv")
```

```python
# Function to summarize with descriptive statistics and a correlation matrix
def summarize_recipes(df):
    print("Summary Statistics")
    display(df.describe())
    
    print("\nCorrelation Matrix")
    display(df[['cost', 'cost_per_serving', 'prep_time', 'cook_time', 'total_time', 'servings']].corr())
```

```python
# All recipes in dataset
summarize_recipes(recipes)
```

```python
# Quick recipes (total_time <= 30 minutes)
quick_recipes = recipes[recipes['total_time'] <= 30]
summarize_recipes(quick_recipes)
```

```python
# Cheap recipes (cost_per_serving <= $2)
cheap_recipes = recipes[recipes['cost_per_serving'] <= 2]
summarize_recipes(cheap_recipes)
```

```python
# Chicken recipes
chicken_recipes = recipes[recipes['title'].str.contains('chicken', case=False)]
summarize_recipes(chicken_recipes)
```

```python
# Beef recipes
beef_recipes = recipes[recipes['title'].str.contains('beef', case=False)]
summarize_recipes(beef_recipes)
```

```python
# One-pot recipes
one_pot_recipes = recipes[recipes['title'].str.contains('one pot|one-pan|one-pot', case=False)]
summarize_recipes(one_pot_recipes)
```

```python
comparison = pd.DataFrame({
    "All Recipes": recipes[['cost','cost_per_serving','total_time']].mean(),
    "Quick Recipes": quick_recipes[['cost','cost_per_serving','total_time']].mean(),
    "Cheap Recipes": cheap_recipes[['cost','cost_per_serving','total_time']].mean()
})

comparison
```
